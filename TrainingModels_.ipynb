```python
import numpy as np
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.rand(100,1)
```

```python
# create a graph with X and y
import matplotlib.pyplot as plt
plot = plt.scatter(X,y)
plt.show()
```

```python
# Compute theta using normal Eqaution
# The function that we used to generate the data is y = 4 + 3x + Gaussian
# noise. Let’s see what the equation found:

X_b = np.c_[np.ones((100, 1)),X] # Add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta_best)
```

```python
# Now yuocan make predictions using theta
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new] # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
print(y_predict)
```

```python
# Plot this model predications
plot = plt.scatter(X,y)
plt.plot(X_new,y_predict,'r-')
plt.axis([0,2,0,15])
plt.show()
```

```python
# Using scikit learn
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
print(lin_reg.intercept_,lin_reg.coef_)
print(lin_reg.predict(X_new))
```

```python
# Implementation of Gradient Descent Function

eta = 0.1 # Learning step
n_iterations = 1000
m = 100

theta = np.random.rand(2,1)
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
print(theta)
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1,solver='cholesky')
ridge_reg.fit(X,y)
print(ridge_reg.predict(X_new))
```

```python
# Using stochastic Gradient Descent to use Rigse
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(penalty='l2')
sgd_reg.fit(X,y.ravel())
print(sgd_reg.predict(X_new))

## l2 shows we want to use ridge
```
